#### ndarray(多維陣列)

原生的Python的容器限制，無法做到向量運算

例如 x = [1, 2, 3] 

如果預期要將x的內容全部都乘2

但將x * 2時執行的結果

會是[1, 2, 3, 1, 2, 3]

而不是 [2, 4, 6]

````python
x = [1, 2, 3]
y = [4, 5, 6]
print(x * 2)
````

因此就要透過numpy 先將list轉換成多維陣列

再將陣列進行資料處理

````python
x1 = np.array(x)
y1 = np.array(y)
````

````python
type(x1)
````

````python
x1
````

````python
y1
````

````python
x1 + y1
````
