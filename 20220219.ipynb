 ## Request

 從政府資料開放平台抓資料：https://data.gov.tw/

抓取年度國內主要金融指標

https://apiservice.mol.gov.tw/OdService/download/A17000000J-030244-dpp

````python
import requests
````

````python
url = 'https://apiservice.mol.gov.tw/OdService/download/A17000000J-030244-dpp'
````

````python
data = requests.get(url).json()
````

````python
data
````

````python
# 檢視data資料型態
type(data)
````

````python
# 檢視data資料有幾筆
len(data)
````

#### 列出貨幣供給額期底年增率-M2

````python
# 方法一
for i in data:
  print(i['年度'],'貨幣供給額期底年增率-M2:', i['貨幣供給額期底年增率-M2'])
````

````python
# 方法二
for i in range(len(data)):
  print(data[i]['年度'],'貨幣供給額期底年增率-M2:', data[i]['貨幣供給額期底年增率-M2'])
````
