## 練習-關於 00878 定期定額報酬率評估

```python
HDY = yf.download('00878.tw')
```

```python
HDY
```

```python
cash = 50000
```

```python
for i in range(12):
  print(f'2022-{i+1:02}')
```

```python
# 總共花的錢
total_cash = cash*12
```

```python
# 總共買的股數
TOTALSTOCK = 0
for i in range(12):
  d = f"2022-{i+1:02}"
  TOTALSTOCK += cash/HDY.loc[d]['Close'][0]
```

```python
TOTALSTOCK
```

```python
# 2022年底結算總價值/花費的金額
TOTALSTOCK*HDY.loc['2022']['Close'][-1]/total_cash
```

```python
HDY.loc['2022']['Close'].plot()
```

### 如果不是定期定額，報酬率會是？

```python
# 在2022年第一天買了幾張股票
LUMPSUM = total_cash/HDY.loc['2022']['Close'][0]
```

```python
# 在2022年年底買掉的投資報酬率
LUMPSUM*HDY.loc['2022']['Close'][-1]/total_cash
```

## 練習-爬取
https://victorgau.github.io/khpy_web_intro/spy.html

```python
url = "https://victorgau.github.io/khpy_web_intro/spy.html"
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get(url).text
soup = BeautifulSoup(r)
```

```python
soup.find("table")
```

```python
table = soup.find("table")
print("表格內容：")
for tr in table.find_all("tr"):
    row = ""
    for td in tr.find_all("td"):
        row += td.text.strip() + "\t"
    print(row)
```
